```python
#1

class Calculator:
    def __init__(self, first_value, operator, second_value):
        self.first_value = first_value
        self.operator = operator
        self.second_value = second_value
        print("Let's Calculate!")
        print(f'Value 1: {self.first_value}')
        print(f'Operator: {self.operator}')
        print(f'Value 2: {self.second_value}')

    def add(self, v1, v2):
        r = v1 + v2
        print(f'Result: {r}')

    def subtract(self, v1, v2):
        r = v1 - v2
        print(f'Result: {r}')

    def multiply(self, v1, v2):
        r = v1 * v2
        print(f'Result: {r}')

    def divide(self, v1, v2):
        r = v1 / v2
        print(f'Result: {r}')


value1 = int(input())
op = input()
value2 = int(input())
m_homework = Calculator(value1, op, value2)
if op == "+":
    m_homework.add(value1, value2)
elif op == "-":
    m_homework.subtract(value1, value2)
elif op == "*":
    m_homework.multiply(value1, value2)
elif op == "/":
    m_homework.divide(value1, value2)
```

```python
#3

class Patient:
    def __init__(self, name, age, weight, height):
        self.name = name
        self.age = age
        self.weight = weight
        self.height = height

    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print(f'Weight: {self.weight}')
        print(f'Height: {self.height}')
        height_convo = self.height * .01
        print(f'BMI: {self.weight / height_convo ** 2}')


p1 = Patient("A", 55, 63.0, 158.0)
p1.printDetails()
print("====================")
p2 = Patient("B", 53, 61.0, 149.0)
p2.printDetails()
```

```python
#4

class Vehicle:
    def __init__(self):
        self.x = 0
        self.y = 0

    def print_position(self):
        print((self.x, self.y))

    def moveUp(self):
        self.x += 1

    def moveLeft(self):
        self.y -= 1

    def moveDown(self):
        self.x -= 1

    def moveRight(self):
        self.y += 1


car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#5


class Shape:
    def __init__(self, name, v1, v2):
        self.name = name
        self.v1 = v1
        self.v2 = v2

    def area(self):
        if self.name == "Triangle":
            print(f'Area: {.5 * self.v1 * self.v2}')

        elif self.name == "Square":
            print(f'Area: {self.v1 * self.v2}')

        elif self.name == "Rhombus":
            print(f'Area: {.5 * self.v1 * self.v2}')

        elif self.name == "Rectangle":
            print(f'Area: {self.v1 * self.v2}')

        elif self.name == 'Trapezium':
            print('Area: Shape unknown')


triangle = Shape("Triangle", 10, 25)
triangle.area()
print("==========================")
square = Shape("Square", 10, 10)
square.area()
print("==========================")
rhombus = Shape("Rhombus", 18, 25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle", 15, 30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium", 15, 30)
trapezium.area()
```

```python
#6

class Calculator:
    def __init__(self):
        print("Calculator is ready!")

    def calculate(self, v1, v2, op):
        self.v1 = v1
        self.v2 = v2
        self.op = op

        if self.op == "+":
            val = self.v1 + self.v2
            return val
        elif op == "-":
            val = self.v1 - self.v2
            return val
        elif op == "*":
            val = self.v1 * self.v2
            return val
        elif op == "/":
            val = self.v1 / self.v2
            return val

    def showCalculation(self):
        print(f'{self.v1}{self.op}{self.v2} = {val}')


c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#7

class Student:
    def __init__(self, name, id, dept, cg):
        self.name = name
        self.id = id
        self.dept = dept
        self.cg = cg


    def calculate_CGPA(self):
        course_num = len(self.cg)
        print(course_num)
        cg_count = 0
        for i in self.cg:
            cg_count += (i * 3)
        self.total_cgpa = cg_count / (course_num*3)

    def print_details(self):
        print(f"Name: {self.name}, ID: {self.id}")
        print(f'Department: {self.dept}')
        print(f'CGPA:{self.total_cgpa}')

        if self.total_cgpa > 3.80:
            print("Your academic standing is 'Highest Distinction'")
        elif self.total_cgpa > 3.65:
            print("Your academic standing is 'High Distinction'")
        elif self.total_cgpa > 3.50:
            print("Your academic standing is 'Distinction'")
        elif self.total_cgpa > 2.00:
            print("Your academic standing is 'Satisfactory'")
        elif self.total_cgpa < 2.00:
            print("Sorry, you cannot graduate")


s1 = Student('Dora', '15995599', 'CSE', [4, 3.7, 3.7, 4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7, 1.3, 1.3, 1.3, 1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================")
s3 = Student('Bob', '13311331', 'CSE', [2, 3, 3, 3.7, 2.7, 2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
#8

class Shinobi:
    def __init__(self, name, rank):
        self.name = name
        self.rank = rank
        self.salary = 0
        self.mission = 0

    def changeRank(self, rank):
        self.rank = rank

    def calSalary(self, mission):
        self.mission = mission

        if self.rank == "Genin":
            self.salary = self.mission * 50

        elif self.rank == 'Chunin':
            self.salary = self.mission * 100

        else:
            self.salary = self.mission * 500

    def printInfo(self):
        print(f'Name: {self.name}')
        print(f'Rank: {self.rank}')
        print(f'Number of mission: {self.mission}')
        print(f'Salary: {self.salary}')



naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
#9

class Programmer:
    def __init__(self, name, lan, exp):
        self.name = name
        self.lan = lan
        self.exp = exp
        print('Horray! A new programmer is born')

    def addExp(self, exp):
        self.exp += exp
        print(f'Updating experience of {self.name}')

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f'Language: {self.lan}')
        print(f'Experience: {self.exp} years.')


p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
#10

class UberEats:
    def __init__(self, name, num, loc):
        self.name = name
        self.num = num
        self.loc = loc
        print(f'{self.name}, welcome to UberEats!')

    def add_items(self, item1, item2, cost1, cost2):
        self.item1 = item1
        self.item2 = item2
        self.cost1 = cost1
        self.cost2 = cost2

    def print_order_detail(self):
        print(f'Name: {self.name}, Phone: {self.num}, Address: {self.loc}')
        print(f"Orders: {{'{self.item1}': {self.cost1}, '{self.item2}': {self.cost2}}}")
        print(f'Total Paid Amount: {self.cost1+self.cost2}')


order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```

```python
#11

class Spotify:
    def __init__(self, songs):
        self.songs = songs
        print("Welcome to Spotify!")

    def playing_number(self, no):
        if no <= len(self.songs):
            print('##########################')
            return f'Playing {no} number song for you\nSong name: {self.songs[no-1]}'

        else:
            return f"{no} number song not found. Your playlist has {len(self.songs)} songs only."

    def add_to_playlist(self, *args):
        for i in args:
            self.songs.append(i)





user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
