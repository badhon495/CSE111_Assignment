```python
#task-4

class Account:
    count = 0

    def __init__(self, name, age, occupation, total):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.total = total
        Account.count += 1

    def addMoney(self, add_money):
        self.total += add_money

    def withdrawMoney(self, withdraw_money):
        if withdraw_money > self.total:
            pass
        else:
            self.total -= withdraw_money

    def printDetails(self):
        print(f'Name: {self.name}\nAge: {self.age}\nOccupation: {self.occupation}\nTotal Amount: {self.total}')


print('No of account holders:', Account.count)
print("=========================")
p1 = Account("Abdul", 45, "Service Holder", 500000)
p1.addMoney(300000)
p1.printDetails()
print("=========================")
p2 = Account("Rahim", 55, "Businessman", 700000)
p2.withdrawMoney(700000)
p2.printDetails()
print("=========================")
p3 = Account("Ashraf", 62, "Govt. Officer", 200000)
p3.withdrawMoney(250000)
p3.printDetails()
print("=========================")
print('No of account holders:', Account.count)
```

```python
#task-5

class Smartphone:
    def __init__(self, name = None, ):
        self.name = name
        self.feature = {}

    def addFeature(self,*args):
        if self.name is not None:
            for i in range(0,len(args),2):
                if args[i] not in self.feature:
                    self.feature[args[i]] = [args[i+1]]
                else:
                    self.feature[args[i]].append(args[i+1])
        else:
            print("Feature can not be added without phone name")

    def setName(self, name):
        self.name = name

    def printDetail(self):
        print(f'Phone Name: {self.name}')
        for i in self.feature:
            print(f'{i}: {",".join(self.feature[i])}')





s1 = Smartphone()
print("=================================")
s1.addFeature('Display', '6.1 inch')
print("=================================")
s1.setName('Samsung Note 20')
s1.addFeature('Display', '6.1 inch')
s1.printDetail()
print("=================================")
s2 = Smartphone('Iphone 12 Pro')
s2.addFeature('Display', '6.2 inch')
s2.addFeature('Ram', '6 GB')
print("=================================")
s2.printDetail()
s2.addFeature('Display', 'Amoled panel')
s2.addFeature('Ram', 'DDR5')
print("=================================")
s2.printDetail()
print("=================================")
```

```python
#task-15

class Player:
    database = {}
    playerNo = 0

    def __init__(self, name, team, jerseyNo):
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo

    def __str__(self):
        return "Name:{}\nTeam:{}\nJersey No:{}".format(self.name, self.team, self.jerseyNo)


class FootballPlayer(Player):
    def __init__(self, name, team, jerseyNo, goal, ret="Not yet retired"):
        super().__init__(name, team, jerseyNo)
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo
        self.goal = goal
        Player.playerNo += 1
        self.retariment = ret
        first, second = self.name.split()
        self.id = f'{Player.playerNo}{first[0]}{second[0]}{jerseyNo}'

        Player.database[self.id] = [name, team, jerseyNo, goal, self.retariment]

    def createPlayer(name, team, jerseyNo, goal, retariment="Not yet retired"):
        temp = FootballPlayer(name, team, jerseyNo, goal, retariment)
        return temp

    def __str__(self):
        print(f'Player ID:{self.id}')
        return f'{super().__str__()}\n{self.retariment}'


print("Number of players:", Player.playerNo)
print("Player Database:", Player.database)
print("#################################")
p1 = FootballPlayer("Lionel Messi", "Barcelona", 10, 231)
print("------Details of the player------")
print(p1)
print("#################################")
p2 = FootballPlayer("Cristiano Ronaldo", "Juventus", 7, 215)
print("------Details of the player------")
print(p2)
print("#################################")
p3 = FootballPlayer.createPlayer("Miroslav Klose", "Lazio", 11, 71, "11 Aug,2014")
print("------Details of the player------")
print(p3)
print("#################################")
print("Number of players:", Player.playerNo)
print("Player Database:", Player.database)
```
