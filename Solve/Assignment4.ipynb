```python
#1

class Customer:

    def __init__(self, name):
        self.name = name
        self.item_list = []

    def greet(self, greet_name=None):

        if greet_name is None:
            print(f"Hello!")
        else:
            print(f"Hello {greet_name}!")

    def purchase(self, *items):
        self.item_list += list(items)
        print(f"{self.name}, you purchased {len(self.item_list)} item(s):")
        print("\n".join(self.item_list))


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#7

class Match:
    def __init__(self, name):
        self.score = 0
        self.over = 0
        self.wicket = 0
        self.batting, self.bowling = str(name).split("-")
        print("5..4..3..2..1.. Play !!!")

    def add_run(self, run):
        self.score += run

    def add_over(self, over):
        if over < 5:
            self.over += over
        else:
            print(f'Warning! Cannot add {over} over/s (5 over match)')

    def add_wicket(self, wicket):
        self.wicket += wicket

    def print_scoreboard(self):
        return f'Batting Team: {self.batting}\nBowling Team: {self.bowling}\nTotal Runs: {self.score} Wickets: {self.wicket} Overs: {self.over}'


match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
#9

class Batsman:
    def __init__(self, *player):
        if len(player) < 3:
            self.name = "New Batsman"
            self.run = player[0]
            self.ball = player[1]

        else:
            self.name = player[0]
            self.run = player[1]
            self.ball = player[2]

    def printCareerStatistics(self):
        print(f'Name: {self.name}\nRuns Scored: {self.run} , Balls Faced: {self.ball}')

    def battingStrikeRate(self):
        return f'{(self.run / self.ball) * 100}'

    def setName(self, new_name):
        self.name = new_name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#12

class TaxiLagbe:
    def __init__(self, number, location):
        self.number = number
        self.location = location
        self.passengers = []
        self.fare = 0

    def addPassenger(self, *passengers):
        for p in list(passengers):
            if len(self.passengers) < 4:
                name, fare = str(p).split("_")
                self.passengers.append(name)
                self.fare += int(fare)
                print(f'Dear {name}! Welcome to TaxiLagbe.')
            else:
                print("Taxi Full! No more passengers can be added.")

    def printDetails(self):
        print(
            f'Trip info for Taxi number: {self.number}\nThis taxi can cover only {self.location} area.\nTotal passengers: {len(self.passengers)}\nPassenger lists:')
        print(",".join(self.passengers))
        print(f"Total collected fare: {self.fare} Taka")


taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#13

class Account():
    def __init__(self, name= "Default Account", balance = 0.0):
        self.name = name
        self.balance = balance

    def details(self):
        return f"{self.name}\n{self.balance}"
    
    def withdraw(self, amount):
        temp_balance = self.balance - amount
        if temp_balance < 3071:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or "
                  "less than minimum.")
        else:
            self.balance = temp_balance
            print(f'Withdraw successful! New balance is:\n{self.balance}')


a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah", 400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
#14

class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.cg = 0
        self.grades = {}

    def calculateGPA(self, courses, semester):
        temp_courses = []
        temp_cg = 0
        course_count = 0
        t_dic = {}
        for i in courses:
            course_name, cg = str(i).split(": ")
            temp_cg += float(cg)
            temp_courses.append(course_name)
            course_count += 1
        cg = (temp_cg / course_count)
        if len(str(cg)) > 4:
            if str(cg)[3:4] == "0":
                cg = str(cg)[0:3]
            else:
                pass
        else:
            print(cg)

        t_dic[tuple(temp_courses)] = float(cg)
        self.grades[semester] = t_dic

    def printDetails(self):
        print(f"Name: {self.name}\nID: {self.id}")
        for i in self.grades.keys():
            print(f'Course taken in {i}:')
            for course, cg in self.grades[i].items():
                for i in course:
                    print(i)
                print(cg)


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#22

class Hotel:
    stuff_id = 0
    guest_id = 0

    def __init__(self, hotel_name):
        self.hotel_name = hotel_name
        self.all_stuff = {}
        self.all_guest = {}

    def addStuff(self, stuff_name, age, phone="000"):

        self.stuff_name = stuff_name
        Hotel.stuff_id += 1
        self.stuff_id = Hotel.stuff_id
        print(f'Staff With ID {self.stuff_id} is added')
        self.stuff_age = age
        self.stuff_phone = phone
        self.all_stuff[self.stuff_id] = [self.stuff_name, self.stuff_age, self.stuff_phone]

    def getStuffById(self, id):
        if id == self.stuff_id:
            return f'Staff ID: {self.stuff_id}\nName: {self.stuff_name}\nAge: {self.stuff_age}\nPhone no.: {self.stuff_phone}'
        else:
            pass

    def addGuest(self, name, age, phone):
        self.guest_name = name
        self.guest_age = age
        self.guest_phone = phone
        Hotel.guest_id += 1
        self.guest_id = Hotel.guest_id
        print(f'Guest With ID {self.guest_id} is added')
        self.all_guest[self.guest_id] = self.guest_name, self.guest_age, self.guest_phone

    def getGuestById(self, id):
        if id == self.guest_id:
            return f'Staff ID: {self.guest_id}\nName: {self.guest_name}\nAge: {self.guest_age}\nPhone no.: {self.guest_phone}'
        else:
            pass

    def allstuffs(self):
        print(f'All Staffs:\nNumber of Staff: {len(self.all_stuff)}')
        for id in self.all_stuff.keys():
            name, age, phone = self.all_stuff[id]
            print(f'Staff ID: {id} Name: {name} Age: {age} Phone no.: {phone}')

    def allGuest(self):
        print(f'All Guest:\nNumber of Guest: {len(self.all_guest)}')

        for id in self.all_guest.keys():
            name, age, phone = self.all_guest[id]
            print(f'Staff ID: {id} Name: {name} Age: {age} Phone no.: {phone}')


h = Hotel('Lakeshore')
h.addStuff('Adam', 26)
print('=================================')
print(h.getStuffById(1))
print('=================================')
h.addGuest('Carol', 35, '123')
print('=================================')
print(h.getGuestById(1))
print('=================================')
h.addGuest('Diana', 32, '431')
print('=================================')
print(h.getGuestById(2))
print('=================================')
h.allstuffs()
print('=================================')
h.allGuest()
```

```python
#23

class Author:
    def __init__(self, name=None):
        self.name = name
        self.book_list = {}
        self.book_amount = 0

    def setName(self, name):
        self.name = name

    def addBook(self, book_name, genre):
        if self.name is None:
            print("A book can not be added without author name")

        else:
            self.book_amount += 1
            if genre not in self.book_list:
                self.book_list[genre] = [book_name]
            else:
                self.book_list[genre].append(book_name)

    def printDetail(self):
        print(f'Number of Book(s): {self.book_amount}\nAuthor Name: {self.name}')
        for genre in self.book_list:
            print(f'{genre}: {", ".join(self.book_list[genre])}')


a1 = Author()
print("=================================")
a1.addBook('Ice', 'Science Fiction')
print("=================================")
a1.setName('Anna Kavan')
a1.addBook('Ice', 'Science Fiction')
a1.printDetail()
print("=================================")
a2 = Author('Humayun Ahmed')
a2.addBook('Onnobhubon', 'Science Fiction')
a2.addBook('Megher Upor Bari', 'Horror')
print("=================================")
a2.printDetail()
a2.addBook('Ireena', 'Science Fiction')
print("=================================")
a2.printDetail()
print("=================================")
```

```python
#24

class Hospital:
    def __init__(self, hospital_name):
        self.hospital_name = hospital_name
        self.doc_dic = {}
        self.pat_dic = {}

    def addDoctor(self, doc):
        self.doc_dic[doc.doc[0]] = [doc.doc[2], doc.doc[3]]

    def addPatient(self, pat):
        self.pat_dic[pat.pat[0]] = [pat.pat[2], pat.pat[3], pat.pat[4]]

    def getPatientByID(self, pat_id):
        pat_info = self.pat_dic[pat_id]
        return f"Patient's ID: {pat_id}\nName: {pat_info[0]}\nAge: {pat_info[1]}\nPhone no.: {pat_info[2]}"

    def getDoctorByID(self, doc_id):
        doc_info = self.doc_dic[doc_id]
        return f"Doctor's ID: {doc_id}\nName: {doc_info[0]}\nSpeciality: {doc_info[1]}"

    def allDoctors(self):
        print(f'Number of Doctors: {len(self.doc_dic)}')
        print(self.doc_dic)

    def allPatients(self):
        print(f'Number of Patients: {len(self.pat_dic)}')
        print(self.pat_dic)


class Doctor:
    def __init__(self, *doc):
        self.doc = doc


class Patient:
    def __init__(self, *pat):
        self.pat = pat


h = Hospital("Evercare")
d1 = Doctor("1d", "Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p", "Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient("2p", "Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
#25

class Vaccine:
    def __init__(self, vac_name, vac_location, next_vac):
        self.vac_name = vac_name
        self.vac_location = vac_location
        self.next_vac = next_vac


class Person:

    def __init__(self, name, age, ocu="General Citizen"):
        self.vac_access = Vaccine
        self.person_name = name
        self.person_age = age
        self.person_job = ocu
        self.first_dose = None
        self.sec_dose = None
        self.vac_name = None

    def pushVaccine(self, vac, dose="first dose"):

        if self.vac_name is not None:
            if self.vac_name == vac.vac_name:
                if dose == "2nd Dose":
                    self.sec_dose = "Given"
                    print(f'2nd dose done for {self.person_name}')
                else:
                    self.first_dose = "Given"
                    self.sec_dose = f'Please come after {vac.next_vac} days'
                    print(f'1st dose done for {self.person_name}')
            else:
                print(f'Sorry {self.person_name}, you can’t take 2 different vaccines')
        else:
            if vac.vac_name == "Sinopharm":
                if self.person_age < 25:
                    print(f"Sorry {self.person_name} ,Minimum age for taking vaccines is 25 years now.")
                else:
                    self.vac_name = vac.vac_name
                    self.first_dose = "Given"
                    self.sec_dose = f'Please come after {vac.next_vac} days'
                    print(f"1st dose done for {self.person_name}")

            else:
                self.vac_name = vac.vac_name
                self.first_dose = "Given"
                self.sec_dose = f'Please come after {vac.next_vac} days'
                print(f"1st dose done for {self.person_name}")

    def showDetail(self):
        print(
            f'Name: {self.person_name} Age: {self.person_age} Type: {self.person_job}\nVaccine name: {self.vac_name}\n1st dose: {self.first_dose}\n2nd dose: {self.sec_dose}')


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person('Bob', 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
