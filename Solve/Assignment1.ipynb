```python
#3

u_input = str(input())
count = 0
start_cap = -1
end_cap = -1

for i in u_input:
    count += 1
    if "A" <= i <= "Z" and start_cap == -1:
        start_cap += count
    elif "A" <= i <= "Z" and start_cap != -1:
        end_cap += count

real = u_input[start_cap + 1:end_cap]

if real == "":
    print("BLANK")
else:
    print(real)
```

```python
#5

u_input = str(input())
error = ""
up = True
low = True
digit = True
s_chr = True

for i in u_input:
    if ord(i) in range(65,91):
        up = False
    elif ord(i) in range(97, 123):
        low = False
    elif i in "0" <= i <= "9":
        digit = False
    elif i in ('_', '$', '#', '@'):
        s_chr = False

if up:
    error += "Uppercase character missing, "
if low:
    error += "Lowercase character missing, "
if digit:
    error += "Digit missing, "
if s_chr:
    error += "Special character missing, "

if len(error) != 0:
    print(error[:-2])
elif len(error) == 0:
    print("OK")
```

```python
#4 - st
phone = {0: ' ', 1: '.,?!:', 2: 'ABC', 3: 'DEF', 4: 'GHI', 5: 'JKL', 6: 'MNO',
         7: 'PQRS', 8: 'TUV', 9: 'WXYZ'}

dict1 = {}

for i, j in phone.items():
    value = str(i)
    count = 1
    for word in j:
        dict1[word] = value * count
        count += 1

u_input = input()
press = ''
for word in u_input:
    press += dict1[word.upper()]

print(press)
```

```python
#1

def bmi(u_input):
    condition = ""
    x = str(u_input[1:-1])
    height, weight = x.split(", ")
    real_height = (int(height) / 100) ** 2
    real_bmi = int(weight) / real_height

    if real_bmi < 18.5:
        condition = "Underweight"
    elif 18.5 <= real_bmi <= 24.9:
        condition = "Normal"
    elif 25 <= real_bmi <= 30:
        condition = "Overweight"
    else:
        condition = "Obese"

    return f'Score is {round(real_bmi, 1)}. You are {condition}'


print(bmi(input()))
```

```python
#2

def number(u_input):
    count = 0
    x = str(u_input[1:-1])
    min, max, div = x.split(", ")

    for i in range(int(min), int(max)):
        if i % int(div) == 0:
            count += i

    return count


print(number(input()))
```

```python
#4

def domain(mail, new_domain, old_domain='sheba.xyz'):
    r_mail = ""
    if new_domain in mail:
        print(f'Unchanged: {mail}')
    else:
        for i in range(len(mail)):
            if mail[i] == "@":
                r_mail = mail[0:i + 1]
        print(f'Changed: {str(r_mail) + new_domain}')


domain('bob@sheba.xyz', 'sheba.xyz')
```

```python
#5

def palindrome(u_input):
    new = ""
    for i in u_input:
        if i == " ":
            continue
        else:
            new += i
    if new == new[::-1]:
        print("Palindrome")
    else:
        print("Not a palindrome")


palindrome(str(input()))
```
